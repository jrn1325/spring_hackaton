```python
from sklearn.linear_model import LinearRegression, Ridge, Lasso, ElasticNet
from sklearn.ensemble import RandomForestRegressor, GradientBoostingRegressor

def linear_regression(X_train, y_train, X_test):
    """
    Compute linear regression
    """
    # create a linear regression model
    model = LinearRegression()

    # fit the model to the training data
    model.fit(X_train, y_train)

    # predict the target values for the test data
    y_pred = model.predict(X_test)

    return y_pred

def ridge_regression(X_train, y_train, X_test):
    # create a ridge regression model
    model = Ridge()

    # fit the model to the training data
    model.fit(X_train, y_train)

    # predict the target values for the test data
    y_pred = model.predict(X_test)

    return y_pred

def lasso_regression(X_train, y_train, X_test):
    # create a lasso regression model
    model = Lasso()

    # fit the model to the training data
    model.fit(X_train, y_train)

    # predict the target values for the test data
    y_pred = model.predict(X_test)

    return y_pred

def elasticnet_regression(X_train, y_train, X_test):
    # create a elasticnet regression model
    model = ElasticNet()

    # fit the model to the training data
    model.fit(X_train, y_train)

    # predict the target values for the test data
    y_pred = model.predict(X_test)

    return y_pred

def random_forest_regression(X_train, y_train, X_test):
    # create a random forest regression model
    model = RandomForestRegressor()

    # fit the model to the training data
    model.fit(X_train, y_train)

    # predict the target values for the test data
    y_pred = model.predict(X_test)

    return y_pred

def gradient_boosting_regression(X_train, y_train, X_test,):
    # create a gradient boosting regression model
    model = GradientBoostingRegressor()

    # fit the model to the training data
    model.fit(X_train, y_train)

    # predict the target values for the test data
    y_pred = model.predict(X_test)

    return y_pred
```
